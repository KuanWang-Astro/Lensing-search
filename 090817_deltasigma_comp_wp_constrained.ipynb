```python
rp_ds = f01['rp'][0]
rp_wp = np.logspace(-1, np.log10(500)-1, 20)
rp_wp = (rp_wp[:-1] + rp_wp[1:])/2.
```

```python
f00 = np.load('090417_wp_ggl_Reid_wo.npz')
```

```python
f00.keys()
```

```python
wp_Reid_wo = np.mean(f00['wprp'][:,2:-3],axis=0)
```

```python
ds_Reid_wo = np.mean(f00['deltasigma'],axis=0)
```

```python
wp_lower = wp_Reid_wo*0.75
wp_upper = wp_Reid_wo*1.25
```

```python
filelist = [f01,f02,f03,f04,f05,f06,f07,f08,f09,f10,f11,f12,f13,f14,f15,f16,f17,f18,f19,f20,f21,f22,f23,f24,f25,f26,f27,f28]
```

```python
idxlist = []
```

```python
for i,f in enumerate(filelist):
    for j,wp in enumerate(f['wprp'][:,2:-3]):
        if all(wp>wp_lower) and all(wp<wp_upper):
            idxlist.append(np.array((i,j)))
```

```python
idxlist = np.array(idxlist)
```

```python
for ij in idxlist:
    wp = filelist[ij[0]]['wprp'][ij[1],2:-3]
    ds = filelist[ij[0]]['deltasigma'][ij[1]]
    plt.figure()
    plt.plot(rp_wp[2:-3],wp,'r',label='test parameters')
    plt.plot(rp_wp[2:-3],wp_Reid_wo,'k',label='Reid parameters')
#    plt.xlabel('r/Mpc')
#    plt.ylabel(r'$\Delta\Sigma$')
    plt.loglog()
    plt.legend()
    plt.title('{},{}'.format(ij[0]+1,ij[1]))
    plt.figure()
    plt.plot(rp_ds,ds,'r',label='test parameters')
    plt.plot(rp_ds,ds_Reid_wo,'k',label='Reid parameters')
    plt.plot(rp_ds,ds_Reid_wo*0.7,'b',label='70% Reid parameters')
    plt.plot(rp_ds,ds_Reid_wo*0.85,'g',label='85% Reid parameters')
    plt.loglog()
    plt.legend()
```

```python
for ij in idxlist:
    wp = filelist[ij[0]]['wprp'][ij[1],2:-3]
    ds = filelist[ij[0]]['deltasigma'][ij[1]]
    print ij+np.array((1,0)),ds/ds_Reid_wo
    plt.figure()
    plt.title('f{},{}'.format(ij[0]+1,ij[1]))
    plt.plot(rp_wp[2:-3],wp/wp_Reid_wo,label='wp ratio')
    plt.plot(rp_ds,ds/ds_Reid_wo,label='delta sigma ratio')
    plt.semilogx()
    plt.legend()
    plt.axhline(1,0,1,c='k')
    plt.ylim((0.75,1.25))
```
