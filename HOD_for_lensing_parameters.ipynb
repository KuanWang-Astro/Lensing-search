```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
from HOD import HOD
hod = HOD()
```

```python
Mvir = np.logspace(5,15,101)
```

```python
param_Reid = np.array((0.73,14.09,0.32,13.27,13.004))
```

```python
standard = np.zeros(101)
standard_cen = np.zeros(101)
```

```python
for i,M in enumerate(Mvir):
    standard[i] = hod.ngal_tot(M,param_Reid)
    standard_cen[i] = hod.ngal_cen(M,param_Reid)
```

```python
plt.plot(Mvir,standard,label='all gals')
plt.plot(Mvir,standard_cen,label='centrals')
plt.loglog()
plt.legend()
plt.title('Reid parameters, w/o AB')
```

```python
param_w = np.loadtxt('promising_params.txt')
```

```python
decorated_h = np.zeros((15,101))
decorated_cen_h = np.zeros((15,101))
decorated_l = np.zeros((15,101))
decorated_cen_l = np.zeros((15,101))
```

```python
for i,param in enumerate(param_w):
    for j,M in enumerate(Mvir):
        decorated_h[i,j] = hod.ngal_tot(M,param)[0]
        decorated_cen_h[i,j] = hod.ngal_cen(M,param)[0]
        decorated_l[i,j] = hod.ngal_tot(M,param)[1]
        decorated_cen_l[i,j] = hod.ngal_cen(M,param)[1]
```

```python
for i in range(15):
    plt.plot(Mvir,decorated_l[i],'b',alpha=0.3)
    plt.plot(Mvir,decorated_cen_l[i],'orange',alpha=0.3)
    plt.plot(Mvir,decorated_h[i],'g',alpha=0.3)
    plt.plot(Mvir,decorated_cen_h[i],'r',alpha=0.3)
    plt.loglog()
    plt.title('lensing parameters, w/ AB')
```
