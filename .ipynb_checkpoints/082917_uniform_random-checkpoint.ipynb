```python
lower = np.array((0,10.7,0.02,9,9,-1,-1))
upper = np.array((2,15,1.5,14,14,1,1))
diff = upper-lower
```

```python
def relative_p(param):
    return (param-lower)/diff
```

```python
for param in param1:
    plt.plot(relative_p(param),range(1,8))
for param in param2:
    plt.plot(relative_p(param),range(1,8))
for param in param3:
    plt.plot(relative_p(param),range(1,8))
    
plt.yticks((1,2,3,4,5,6,7),('alpha','logM1','sigma_logM','logM0','logMmin','Acen','Asat'))
plt.xticks((0,1),('lower','upper'))
```

```python
stat0 = np.load('090417_wp_ggl_Reid_wo.npz')
stat00 = np.load('090417_wp_ggl_Reid_fiducial_wo.npz')
stat00_80 = np.load('090417_wp_ggl_Reid_fiducial_wo_pimax80.npz')
```

```python
wp0 = stat0['wprp']
ds0 = stat0['deltasigma']
wp00 = stat00['wprp']
ds00 = stat00['deltasigma']
wp00_80 = stat00_80['wprp']
ds00_80 = stat00_80['deltasigma']
```

```python
rp_ds = stat0['rp'][0]
rp_wp = np.logspace(-1, np.log10(500)-1, 20)
r_wp = (rp_wp[:-1] + rp_wp[1:])/2.
```

```python
def mean_16_84(name):
    length = name[0].size
    mean = np.zeros(length)
    percentile16 = np.zeros(length)
    percentile84 = np.zeros(length)

    for i in range(length):
        mean[i] = np.mean(name[:,i])
        percentile16[i] = np.percentile(name[:,i],16)
        percentile84[i] = np.percentile(name[:,i],84)

    return mean, percentile16, percentile84
```

```python
wp0_mean = mean_16_84(wp0)[0]
ds0_mean = mean_16_84(ds0)[0]
wp00_mean = mean_16_84(wp00)[0]
ds00_mean = mean_16_84(ds00)[0]
wp00_80_mean = mean_16_84(wp00_80)[0]
ds00_80_mean = mean_16_84(ds00_80)[0]
```
